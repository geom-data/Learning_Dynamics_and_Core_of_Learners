```python
import pandas as pd
```

```python
df = pd.read_csv('./single_models/master_results.csv')
```

```python
group_columns = [col for col in df.columns if col not in ['testing_model_idx', 'acc', 'loss']]
test_df = df.groupby(group_columns).agg({'acc': 'mean', 'loss': 'mean'}).reset_index()
test_df = test_df.rename(columns={'acc': 'avg_acc', 'loss': 'avg_loss'})
```

```python
only_B_df_aggregated.sort_values(by=['D_generating_model','D_attack_direction', 'D_attack_type'], inplace=True)
```

```python
only_B_df_aggregated.reset_index(drop=True, inplace = True)
```

```python
only_B_df_aggregated.drop(columns=['testing_model_augmentation'], inplace=True)
only_B_df_aggregated
```
